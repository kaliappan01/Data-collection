```python
import requests
import json
def state_district_wise_cases(state,district,type):
  response_API = requests.get('https://api.covid19india.org/state_district_wise.json')
  data = response_API.text
  parse_json = json.loads(data)
  active_case = parse_json[state]['districtData'][district][type]
  print("Active cases in {} : {} is {}".format(state,district,active_case))
```

```python
state_district_wise_cases("Maharashtra","Raigad","active")
```
